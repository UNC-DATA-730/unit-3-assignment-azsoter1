- `lstat`: lower status of the population (percent). This seems to indicate the percentage of the population classified as being "lower status".
- `medv`: median value of owner-occupied homes in $1000s.

Let's plot `medv` (y-axis) versus `lstat`.

**Answer:**

The relationship appear to be negative: in the suburbs where the percentage of the population classified as being "lower status" is higher the median value of owner-occupied homes is usually lower.

### Let's build a model!

We need to first load the [Tidymodels]() package.

````r
library('tidymodels')
````

First we specify our model as `linear regression` using the `lm` engine.

````r
mod = linear_reg() |> set_engine("lm")

mod
````

Next we "fit" our model by supplying the `formula` and the data.

````r
mod_fit = mod |> fit(medv ~ lstat, data = Boston)

mod_fit
````

### Get the regression table

````r
# We can use the tidy function to get a table of our model information
tidy(mod_fit)
````

❓Comparing on the chart above and the regression table, does the **sign** of the `estimate` for the `lstat` term coefficient fit your expectations?

**Answer:**

Yes, the sign of the estimate for the lstat term (which is the slope of the regression line) is negative, which was in line with my earlier expectation (where I also saw a negative relationship in the scatterplot chart). The slope of the regression line (-0.95) is different from the correlation coefficient I estimated earlier by taking a look at the scatterplot (my correlation coefficient estimate was in the range of -0.6 to -0.8). This is not surprising: while the correlation coefficient and the slope of a regression line always have the same sign, they typically do not have the same value.

### Making predictions and visualizing the model

We can use the `augment` function to "predict" `medv` for all the values in our original dataset. We will capture these predictions in a new data frame called `Boston2`. The predicted values are found in the `.pred` column.

````r
Boston2 = augment(mod_fit, Boston)

glimpse(Boston2)
````

Let's visualize our model.

````r
p = ggplot(Boston2, aes(x = lstat)) +
    geom_point(aes(y = medv)) + 
    geom_line(aes(y = .pred), color = 'coral', linewidth = 1.5)

p
````

Yes, looking at the chart, it seems that the stores with good shelf location have the highest number of units of car seats sold in general followed by stores with medium shelf location and stores with bad shelf location. In cells below, using the `Carseats` data build a model of `Sales` with the `ShelveLoc` variable as a predictor.

**Include the following:**
- Show a regression table of your model parameters.
- Visualize the model with `ggplot2`.

📊 *Here is some example code for plotting your model.*

```r
# The code below assumes your predictions column is called ".pred" and is in a
# data frame called "Carseats2"

ggplot(Carseats2, aes(x = ShelveLoc)) + 
    geom_point(aes(y = Sales), 
               na.rm = T, position = position_jitter(height = 0, width = 0.2, seed = 42)) +
    geom_crossbar(aes(y = .pred, ymin = .pred, ymax = .pred), 
                  color = 'coral')
```

````r
mod3 = linear_reg() |> set_engine('lm')

mod3_fit = mod3 |> fit(Sales ~ ShelveLoc, data = Carseats)

tidy(mod3_fit)
````

````r
Carseats2 = augment(mod3_fit, Carseats)
glimpse(Carseats2)
````

````r
ggplot(Carseats2, aes(x = ShelveLoc)) + 
    geom_point(aes(y = Sales), 
               na.rm = TRUE, position = position_jitter(height = 0, width = 0.2, seed = 42)) +
    geom_segment(aes(xend = ShelveLoc, y = Sales, yend = .pred), 
                 na.rm = TRUE, position = position_jitter(height = 0, width = 0.2, seed = 42), 
                 color = 'steelblue',
                 alpha = 0.25) +
    geom_crossbar(aes(y = .pred, ymin = .pred, ymax = .pred),
                  color = 'coral')
````

Which `ShelveLoc` category does your intercept term represent? 

**Answer:**

In case of linear models with a categorical variable the intercept term in the regression table represents the first category member in aphabetical order. As a result, in this case the intercept term represents the "Bad" ShelveLoc category.
